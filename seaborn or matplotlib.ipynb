```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

data ={
    'student':['A', 'B','C','D','E'],
    'math':[85,78,89,65,90],
    'science':[80,75,89,65,90],
    'english':[78,82,88,72,85]
}
df= pd.DataFrame(data)
print(df)
```

```python
sns.histplot(df['math'], kde =True, color ='orange')

plt.title("Histogram of math scores")
plt.show()
```

```python
sns.scatterplot(x='math',y='science', data=df)

plt.title("Math vs science scores")
plt.xlabel("math score")
plt.ylabel("science score")
plt.show()
```

```python
corr =df.drop('student',axis=1).corr() #corr(): pairwise correlation of columns within a DataFrame

sns.heatmap(corr, annot=True,cmap='YlGnBu')

plt.title("subject score correlation")
plt.show()
```
